# Vector data

#### In Geographic Information Systems (GIS), vector data is a type of spatial data representation used to represent discrete geographic features as points, lines, and polygons. It is a fundamental data model that allows for the storage, analysis, and visualization of geographic information.

#### Vector data represents real-world features, such as roads, buildings, rivers, boundaries, and points of interest, as discrete objects with well-defined boundaries and attributes. Here are the three main types of vector data:

### 1) Points: 
#### Points represent individual locations or specific points of interest, such as the location of a city, a landmark, or a GPS coordinate.

### 2) Lines (or polylines): 
#### Lines represent linear features, such as roads, rivers, pipelines, or boundaries. They are made up of a sequence of connected vertices, forming a path.

### 3) Polygons: 
#### Polygons represent enclosed areas or regions, such as land parcels, lakes, administrative boundaries, or buildings. They are defined by a series of interconnected vertices, forming a closed shape.

#### In addition to the geometric representation, vector data also includes attribute data associated with each feature. Attributes provide additional information about the geographic features, such as the name, population, elevation, land use, or any other descriptive information.

#### Vector data is typically stored in file formats such as shapefiles (.shp), GeoJSON (.geojson), or Geodatabase (.gdb), and can be created, edited, and analyzed using GIS software.

## Shapefile

#### A shapefile is a popular geospatial vector data format used in Geographic Information Systems (GIS). It consists of a collection of files that work together to store and represent geographic features and attribute data. The shapefile format was developed by Esri, the leading GIS software provider.

#### A shapefile is composed of several individual files with specific extensions, which collectively make up a complete shapefile dataset. The main files that constitute a shapefile dataset are:

#### 1. .shp: The shapefile itself stores the geometric data, such as points, lines, or polygons, that represent the geographic features. It contains the coordinates and topology information of the features.

#### 2. .shx: The shape index file is an index file that allows for efficient random access to the records in the shapefile. It helps in quickly locating specific features within the shapefile.

#### 3. .dbf: The dBASE table file stores the attribute data associated with each geographic feature. It is a simple database file that can contain multiple fields and records, with each record representing a feature and each field representing a specific attribute.

#### Optional supporting files that may accompany a shapefile dataset include:

#### 4. .prj: The projection file specifies the coordinate system and projection information of the shapefile. It defines how the geographic features are aligned to a specific spatial reference system.

#### 5. .sbn and .sbx: These are spatial index files that improve the spatial querying performance of shapefiles. They create a spatial index structure for faster access to features based on their spatial location.

#### 6. .shp.xml: An optional XML metadata file that can provide additional information about the shapefile, such as the data source, author, description, and other metadata elements.

#### These files, collectively organized in the same directory and sharing the same base filename, form a shapefile dataset. All the files must be present and have the same base filename (e.g., "filename.shp", "filename.shx", "filename.dbf") for the shapefile to be correctly read and interpreted by GIS software.

#### Shapefiles are widely supported by various GIS software and can store both simple and complex geographic features, making them a widely used and versatile format for storing and exchanging spatial data.

# Import libraries

```python
# import libraries
import matplotlib.pyplot as plt
import geopandas as gpd
import contextily as cx
```

# Plot Single Shapefile

```python
# Read the shapefile
data = gpd.read_file('E:/Python/Geospatial python/data/vector/point.shp')

# define figure dimension and axes
fig, ax = plt.subplots(figsize=(8, 8))

# plot the data
data.plot(ax=ax, color='red', label='Point')

# add title, xlabel, ylabel, legend
plt.title('Point data')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Legend', loc='lower right')

# Add basemap using contextily (using OpenStreetMap)
cx.add_basemap(ax, source=cx.providers.OpenStreetMap.Mapnik, crs=data.crs)

# Show the plot
plt.show()
```

# Plot Multiple Shapefiles

```python
# Read the shapefiles
point = gpd.read_file('E:/Python/Geospatial python/data/vector/point.shp')
line = gpd.read_file('E:/Python/Geospatial python/data/vector/line.shp')
polygon = gpd.read_file('E:/Python/Geospatial python/data/vector/polygon.shp')

# Create a new plot
fig, ax = plt.subplots(figsize=(10, 10))

# Plot the shapefiles
point.plot(ax=ax, color='red', label='Point', marker='o', markersize=20)
line.plot(ax=ax, color='blue', label='Line', linewidth=2)
polygon.plot(ax=ax, color='green', label='Polygon')

# Customize the plot
ax.set_title('Vector Data Plot')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

import matplotlib.lines as mlines
# Create custom legend handles
point_legend = mlines.Line2D([], [], color='red', marker='o', linestyle='', markersize=10, label='Point')
line_legend = mlines.Line2D([], [], color='blue', linewidth=2, label='Line')

# import mpatches from matplotlib
import matplotlib.patches as mpatches
polygon = mpatches.Patch(color='green', alpha=0.75, label='Polygon')

# Add legend
plt.legend(handles=[point_legend, line_legend, polygon],
    #bbox_to_anchor=(0.99, 0.2),
    loc='lower right',
    borderaxespad=1,
    title='Legend',
    frameon=True,
    prop={'size': 12})

# Add north arrow
arrow_props = dict(facecolor='black', arrowstyle='->')
ax.annotate('N', xy=(0.96, 0.96), xycoords='axes fraction', ha='center', va='center', fontsize=12, xytext=(0.96, 0.92),
            textcoords='axes fraction', arrowprops=arrow_props)

# Add basemap using contextily (using OpenStreetMap)
cx.add_basemap(ax, source=cx.providers.OpenStreetMap.Mapnik, crs=data.crs)

# Add cartoDB basemap
# cx.add_basemap(ax, source=cx.providers.CartoDB.Positron, crs=data.crs)

# check https://contextily.readthedocs.io/en/latest/providers_deepdive.html for more basemaps

# Display the plot
plt.show()
```
